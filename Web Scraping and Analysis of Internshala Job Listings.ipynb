```python
Company = []
Role = []
Location = []
Join = []
Salary=[]
min_Experience = []
max_Experience = []
Posted = []

for i in range(1,12):
    url =f"https://internshala.com/jobs/matching-preferences/page-{i}/"
    page=requests.get(url)
    soup=BeautifulSoup(page.text)



    for i in soup.find_all("div",class_="internship_meta"):
        company= i.find("a",class_="link_display_like_text view_detail_button").text
        if company:
            regex= re.findall("(\w+\s?\w+?\s?\w+),?",company)
            if regex:
                Company.append(regex[0])
        else:
            Company.append(np.nan)
            
            
    
    for i in soup.find_all("div",class_="internship_meta"):
        role = i.find("a",class_="view_detail_button").text
        if role:
            Role.append(role)
        else:
            Role.append(np.nan)
        
          
        

    for i in soup.find_all("div",class_="internship_meta"):
        location= i.find("a",class_="location_link view_detail_button").text
        if location:
            regex= re.findall("(\w+\s?\w+?\s?\w+),?",location)
            if regex:
                Location.append(regex[0])
        else:
            Location.append(np.nan)



    for i in soup.find_all("div",class_="internship_meta"):
        join= i.find("div",class_="item_body").text
        if join:
            regex= re.findall("(\w+\s?\w+?\s?\w+),?",join)
            if regex:
                Join.append(regex[0])
        else:
            Join.append(np.nan)



    for i in soup.find_all("div",class_="internship_meta"):
        salary= i.find("span",class_="mobile").text


        regex=re.findall(r' ([\d,]+)',salary)
        if regex:
            Salary.append(regex[0])
        else:
            Salary.append(np.nan)



    for i in soup.find_all("div",class_="internship_meta"):
        experience= i.find("div",class_="item_body desktop-text").text
        if experience:
            regex= re.findall("\w|.+",experience)
            if regex:
                min_Experience.append(regex[0])
        else:
            min_Experience.append(np.nan)



    for i in soup.find_all("div",class_="internship_meta"):
        experience= i.find("div",class_="item_body mobile-text").text
        if experience:
            regex= re.findall("\d\s",experience)
            if regex:
                max_Experience.append(regex[0])
        else:
            max_Experience.append(np.nan)




    for i in soup.find_all("div",class_="internship_meta"):
        posted= i.find("div",class_="posted_by_container").text
        if posted:
            regex= re.findall("(\w+\s?\w+?\s?\w+),?",posted)
            if regex:
                Posted.append(regex[0])
        else:
            Posted.append(np.nan)


```
